## **Description:**

This project explores the transition between Impressionism and Cubism
using data analysis and visualization.


It focuses on:
- Temporal evolution
- Artist distribution
- Comparative artistic output


### **Imports**

```python
import pandas as pd
import plotly.express as px
```

### **Dataset**

```python
data = [
    # Impressionists
    ["Claude Monet", "France", 1872],
    ["Claude Monet", "France", 1880],
    ["Edgar Degas", "France", 1875],
    ["Pierre-Auguste Renoir", "France", 1876],
    ["Camille Pissarro", "Denmark", 1874],
    ["Berthe Morisot", "France", 1878],

    # Cubists
    ["Pablo Picasso", "Spain", 1907],
    ["Pablo Picasso", "Spain", 1912],
    ["Georges Braque", "France", 1908],
    ["Juan Gris", "Spain", 1915],
    ["Fernand Léger", "France", 1913],
    ["Robert Delaunay", "France", 1912],
]

df = pd.DataFrame(data, columns=["artist", "country", "year"])

print(df)
```

### **Movement classification**

```python
impressionists = [
    "Claude Monet", "Edgar Degas", "Pierre-Auguste Renoir",
    "Camille Pissarro", "Berthe Morisot"
]

cubists = [
    "Pablo Picasso", "Georges Braque", "Juan Gris",
    "Fernand Léger", "Robert Delaunay"
]

def assign_movement(artist):
    if artist in impressionists:
        return "Impressionism"
    elif artist in cubists:
        return "Cubism"
    else:
        return "Other"

df["movement"] = df["artist"].apply(assign_movement)

print(df)
```

### **Data cleaning**

```python
df.isnull().sum()
```

```python
df.duplicated()
```

```python
df = df.dropna()

df["year"] = df["year"].astype(int)

print(df)
```

### **Timeline analysis**

```python
timeline = df.groupby(["year", "movement"]).size().reset_index(name="count")

fig = px.line(
    timeline,
    x="year",
    y="count",
    color="movement",
    title="Temporal Evolution of Impressionism and Cubism"
)

fig.show()
```

### **Geographical Analysis**

```python
geo = df.groupby(["country", "movement"]).size().reset_index(name="count")

fig = px.bar(
    geo,
    x="country",
    y="count",
    color="movement",
    title="Geographical Distribution of Artists"
)

fig.show()
```

### **Artistic output comparison**

```python
artist_output = df.groupby(["artist", "movement"]).size().reset_index(name="count")

fig = px.bar(
    artist_output,
    x="artist",
    y="count",
    color="movement",
    title="Artistic Output per Artist"
)

fig.show()
```

### **Distribution**

```python
fig = px.box(
    df,
    x="movement",
    y="year",
    title="Distribution of Artwork Years"
)

fig.show()
```

## **Key Insights**

-----


- Impressionism and Cubism overlap slightly in time,
  suggesting a gradual transition rather than a sharp break.

- France appears as a central hub for both movements.

- Cubist artists show a tendency toward later and more
  concentrated production periods.

### **Limitations**

----
- Small dataset
- Movement classification is manually assigned
- Data does not represent full artistic production
